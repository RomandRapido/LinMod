```r
library(tidyverse)
library(ggplot2)
library(dplyr)
```

```r
df <- read_csv("internet.csv")

head(df)
```

```r
model = lm(df$Int ~ df$Gdp)

model
```

```r
coefs = coef(model)
```

```r
coefs[2]
```

```r
formula <- function(X, coefs){
    return (coefs[2]*X + coefs[1])
}

df$expected = formula(df$Gdp, coefs)
```

```r
df$error = df$expected - df$Gdp

df
```

```r
summary(model)
```

```r
plot(model)
```

```r
options(repr.plot.width=12, repr.plot.height=8)
ggplot(df) + 
geom_point(color="blue",mapping=aes(x=Gdp, y=Int)) +
geom_line(mapping=aes(x=Gdp, expected, color="Model")) +
scale_color_manual(values = c("Model" = "pink")) +
labs(title = "Gdp v Int", x="Gdp [Per Thousand]", y="Internet [Perc]") + 
theme(
    plot.title = element_text(size=20, hjust=0.5),
    axis.title = element_text(size=15),
    legend.text = element_text(size=20)
)
```
